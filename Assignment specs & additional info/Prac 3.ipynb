# Retrieving Data

```python
abalone_p = 'abalone.data.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
abalone = pd.read_csv(abalone_p, sep=',', decimal='.',
header=None, names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight',
'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings'])
```

# Manipulating the Data

```python
abalone['Sex'][3]
```

```python
abalone.loc[3]['Sex']
```

```python
abalone.iloc[3]['Sex']
```

```python
abalone.iloc[3][0]
```

```python
mask_ring21 = abalone['Rings'] == 21
```

```python
abalone.loc[mask_ring21,'Diameter']
```

```python
abalone.groupby(['Rings']).mean()
```

```python
abalone.sort_values(by='Length').head()
```
